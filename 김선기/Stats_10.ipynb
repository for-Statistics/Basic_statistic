```python
import pandas as pd
import numpy as np

url = 'https://gist.githubusercontent.com/euphoris/f8cc761601d81d8ca077a9fcacf355b5/raw/46f23ba0efdbf315fcf9536b83a7c417590cd259/crab.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
from statsmodels.formula.api import ols

model = ols('y ~ sat + weight + width', df) #모형을 만든다.
```

```python
res = model.fit() #피팅
```

```python
res.summary() #분석 결과를 확인.
```

sat 와 width는 유의수준 5%에서 유의미하고, weight는 통계적으로 유의미하지 않는다.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

model.exog_names
```

모형식에서 독립변수는 절편, sat, weight, width 순

```python
pd.DataFrame({'컬럼': column, 'VIF': variance_inflation_factor(model.exog, i)} 
             for i, column in enumerate(model.exog_names)
             if column != 'Intercept')  
```

weight 와 width 의 VIF가 각각 4.8, 4.6 이므로, 게의 무게와 너비는 서로 상관이 높아 VIF가 높게 측정되었음.
