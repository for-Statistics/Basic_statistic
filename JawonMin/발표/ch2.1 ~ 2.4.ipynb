### 2.1.3. 표본추출방법

###### (1) 임의표본추출
 - 임의(Random, 무작위)은 <b>대표성</b>을 담보하기 위한 핵심
 - 임의표본추출을 잘 하기 위해서는 모집단 정의가 매우 중요
 - 이를 통해 얻은 표본을 '단순임의표본'이라고 함

###### (2) 층화표본추출
 - 모집단을 동질적인 몇 개의 층(strata)으로 나누고, 각 층에서 단순 무작위 표본추출
 - *ex) 전국 고등학생의 평균 키를 얻기 위해, 각 학년별로 표본을 추출하는 것*
 
###### (3) 군집표본추출(=집락표본추출)
 - 각 집단이 모집단의 구성요소를 대표할 수 있는 이질적 요소로 구성되며 집락과 집락들 사이의 차이가 적은 경우
 - *ex) 전국 고등학생의 평균 키를 얻기 위해, 이과에서 몇명, 문과에서 몇명을 임의추출 *

###### 층화추출 vs 군집추출
 - 층화: 집단내 동질적, 집단간 이질적
 - 군집: 집단내 이질적, 집단간 동질적

### 2.3.2 표준오차(SE, Standard Error)
 - 표본평균의 표준편차로 표본통계량의 변동성을 나타내는 지표
 - 표본값들의 표준편차 $s$와 표본크기 $n$을 기반으로 추정할 수 있다. (중심극한정리, 즉 통계 이론에 근거한 추정)
 $$ SE = \frac {s} {\sqrt{n}} $$
 
 - 데이터 과학에서는 부트스트랩을 통해 표준편차를 추정할 수 있다.

### 2.3.1 중심극한정리

###### 중심극한정리(CLT, Central Limit Theorem)
 - 표본의 크기가 크면, 모집단(평균이 $\mu$, 표준편차가 $\sigma$)의 분포와 상관 없이 표본 평균의 분포는 평균이 $\mu$, 표준편차가 $\frac {\sigma} {\sqrt{n}}$인 정규분포를 따른다.
 
###### 통계학에서 중심극한정리가 중요한 이유
 - 모집단의 분포를 몰라도 확률값을 계산할 수 있는 근거가 되기 때문
 - 하지만 데이터 과학의 경우 부트스트랩을 사용할 수 있기 때문에 통계학 관점보다 덜 중요

## 2.4 부트스트랩

표본분포에 대한 정보, 즉 표본 변동성에 대한 정보를 주는 것이다.

###### 부트스트랩이란?
 - 표본통계량의 변동성을 평가하는 강력한 도구 (*NOT* 표본 크기를 늘이는 도구)
 - 우리가 가진 표본에서 복원추출하여 크기가 n인 또다른 표본을 R개를 얻는다. R개 표본에서 얻은 통계량의 변동을 살펴보는 것
 
 
 - 이를 통해 원래 갖고있던 표본으로부터 얻은 통계량이 어느 정도의 변동을 갖는지 이해할 수 있다.

```python
# 복원추출하여 크기 100인 리샘플 만들기
def bootstrap(sample_size, num_of_resample):
    # 재표본추출된 표본집단의 통계량(평균)을 저장할 list
    statistics = [] 
    
    for i in range(0, num_of_resample):
        # 표본으로부터 크기가 sample_size인 resample 추출하여 통계량(mean) 계산
        resample_mean = data.sample(sample_size, replace=True)['SalePrice'].mean() 
        statistics.append(resample_mean)
    
    return statistics
```

```python
bootstrap_stats = bootstrap(100, 1000)
```

 - 우리는 모집단, 모수에 관심이 있지만, 참값을 알아내는 것은 불가능하다
 - 따라서, 우리는 표본을 통해 추정하려고 한다.
 - 이 때, 추정치가 참값에 얼마나 접근했는지, 즉 추정치를 얼마나 신뢰할 수 있는지를 알아보는 한 가지 방법은
 - 모집단으로부터 표본을 반복해서 뽑고, 각 표본으로부터 계산하여 얼마나 변동하는지를 확인하는 것이다.
 - 하지만 이건 불가능하기 때문에, 모집단의 shape에 대해 가정을 하거나 이미 갖고있는 표본의 정보를 이용해야 한다.
 
 - 예를들어 모집단이 정규분포/베르누이 등 접근하기 쉬운 분포를 따른다고 가정했다고 상상해보자
 - 앞선 전략에 따라, 샘플과 동일한 크기의 샘플을 반복적으로 생성하고, 표본통계량의 변동을 살펴봄으로써, 샘플에 따라 얼마나 달라질 수 있는지 알 수 있다.
 - 이 방법은 합리적이다. 왜냐하면 우리가 가진 표본 중 가장 좋은 표본이고, 실제로 모집단이 어떻게 생겼는지에 대한 유일한 정보이기 때문이다.
 - 또한, 무작위로 선택된 표본들은 모집단과 유사해 보일 것이고, 결과적으로 우리의 것도 그럴 가능성이 높다.
