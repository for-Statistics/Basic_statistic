## 2. 데이터와 표본분포
1. 표본이란
 - 더 큰 데이터 집합에서 얻은 부분집합
 - 모집단으로부터 얻은 부분집합
2. 표본추출의 필요성
 - 모집단을 분석한다는건 시간적, 경제적으로도 힘듦. 현실적으로 어려움
 - 또한, 빅데이터 시대에 데이터의 양이 방대해졌는데 단순히 데이터 양이 많다고 좋은게 아님. 품질이 좋아야함. 
 - 따라서, 양질의 데이터를 수집(?), 추출(?)하는것 즉, 제대로된 표본추출이 필요함

* 전통적인 통계학과 데이터 과학에서 표본, 데이터에 대한 관점의 차이
 - 전통적인 통계학: 가정에 기초한 이론을 통해 모집단을 밝히는 것에 초점
 - 데이터 과학: 가정에 기초하지 않은, 표본에 대한 연구로 방향이 옮겨짐. 따라서, 이론적인 측면보다 표본추출 과정과 데이터에 집중할 필요가 있다. 왜? 표본에 대한 연구를 하는 만큼 그 표본이 양질이어야 하니까.
 
Q. 의문점
 - 설문조사처럼 실제로 표본을 추출하는 현업군이 있겠지만, 유통사 금융사 등은 표본을 추출해서 그 데이터를 분석한다기보다는 이미 수집되어있는 데이터를 활용해서 분석하는 시스템이라고 이해하고 있다.
 - 그 데이터 자체가 표본이기도 하지만, 분석 대상이 될 데이터를 수집할 때 '표본을 추출한다'는 표현보다는 '목적에 맞는 데이터를 정의한다'는 표현이 더 많이 쓰일 것 같다. 
 - '목적에 맞는 데이터를 정의한다'는 것 자체가 우리가 알고자 하는 것(모집단의 특성)을 잘 반영할 수 있는 데이터로 분석해야 한다는 것을 내포하고 있다고 생각하고, 이것이 표본추출의 기본 원리라고 생각한다.
 - 따라서, 일부 현업에서 '데이터를 정의하는 것'이 표본추출과 같은 퍼포먼스가 아닐까 생각하는데 다들 어떻게 생각하는지
